```python
import pandas as pd
```

```python
data=pd.read_csv("Titanic-Dataset.csv")
data
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data["Cabin"].value_counts()
```

```python
data.drop("Cabin",axis=1,inplace=True)
data
```

```python
data["Age"] = data["Age"].fillna(data["Age"].median())
data
```

```python
data.isna().sum()
```

```python
data.dropna(subset=['Embarked'], inplace=True)
```

```python
data.isna().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x="Survived", data=data)
plt.title("Survival Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=data, x="Age", hue="Survived",
             multiple="stack", kde=False, bins=30)
plt.title("Age Distribution by Survival")
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.boxplot(x="Pclass", y="Fare", hue="Survived", data=data)
plt.ylim(0, 150) 
plt.title("Fare Ã— Pclass Ã— Survival")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

data["Sex"] = data["Sex"].map({"male": 0, "female": 1})     

embarked_le = LabelEncoder()
data['Embarked'] = embarked_le.fit_transform(data['Embarked'])

name_encoder = LabelEncoder()
data['Name'] = name_encoder.fit_transform(data['Name'])

ticket_encoder = LabelEncoder()
data['Ticket'] = ticket_encoder.fit_transform(data['Ticket'])

data
```

```python

X=data.drop(["Survived"],axis=1,inplace =False)
y=data["Survived"]
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=44, shuffle =True)

LogisticRegressionModel = LogisticRegression(penalty='l2',solver='sag',C=1.0,random_state=33)
LogisticRegressionModel.fit(X_train, y_train)
```

```python
y_pred = LogisticRegressionModel.predict(X_test)
y_pred
```

```python
y_pred_prob = LogisticRegressionModel.predict_proba(X_test)
y_pred_prob
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix  
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
```
